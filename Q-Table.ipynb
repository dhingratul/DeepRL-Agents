```python
import gym
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

### Load the environment

```python
env = gym.make('FrozenLake-v0')
```

```python
#Initialize table with all zeros
# Q is 16x4 because of 16 possible OBSERVATION SPACE(4x4 grid) and 4 ACTION SPACE at each step(U, D, L ,R)
Q = np.zeros([env.observation_space.n,env.action_space.n])
```

```python
# Set learning parameters
lr = .8
gamma = .95 # Discount
num_episodes = 2000
max_steps = 99 # Per episode
rList = []
```

### Implement Q-Table learning algorithm

```python
for i in range(num_episodes):
    #Reset environment and get first new observation
    s = env.reset()  # State
    rAll = 0  # Total Reward within an episode
    done = False
    #The Q-Table learning algorithm
    for j in range(max_steps):
        #Choose an action by greedily (with noise) picking from Q table
        # (1 / i+1) controls that as the number of episodes increase, so does our confidence in Q values
        a = np.argmax(Q[s,:] + np.random.randn(1,env.action_space.n)*(1./(i+1)))
        
        #Get new state and reward from environment
        new_state, r, done, _ = env.step(a)
        
        #Update Q-Table with new knowledge -- Bellman Equation
        Q[s,a] = Q[s,a] + lr*(r + gamma * np.max(Q[new_state,:]) - Q[s,a])
        rAll += r
        s = new_state
        if done == True:
            break
    #jList.append(j)
    rList.append(rAll)
```

```python
print("Score over time: " +  str(sum(rList)/num_episodes))
```

```python
print("Final Q-Table Values")
print(Q)
```

```python
# Use our Q-table to play FrozenLake
env.reset()

for episode in range(200):
    state = env.reset()
    done = False

    for step in range(max_steps):
        env.render()
        # Take the action (index) that have the maximum expected future reward given that state
        action = np.argmax(Q[state,:])
        
        new_state, reward, done, info = env.step(action)
    
        if done:
            break
        else:
            state = new_state
env.close()
```
